# Korelasi dan *Scatter plot*

Fungsi ini digunakan untuk menampilkan koefisien korelasi antara Customer_Age dengan Months_on_book.

```
cor(BankChurners$Customer_Age , BankChurners$Months_on_book) #Nilai korelasi
```

Fungsi ini digunakan untuk menguji korelasi antara Customer_Age dengan Months_on_book.

```
cor.test(BankChurners$Customer_Age , BankChurners$Months_on_book) #Uji Korelasi
```

Fungsi ini untuk menampilkan *Scatter plot* yang merupakan visualisasi korelasi. Pada kasus ini, *Scatter plot* menampilkan korelasi antara Customer_Age dengan Months_on_book.

```
plot(BankChurners$Customer_Age,BankChurners$Months_on_book, type = "p") #menampilkan scatter plot
```

# Histogram dan Density 

Fungsi ini digunakan untuk menampilkan histogram dan juga density yang merupakan visualisasi frekuensi data. 

```
hist(BankChurners$Customer_Age, main="Umur Customer",col ="light blue") #Histogram

plot(density(BankChurners$Customer_Age),col = "red",main="Umur Customer") #Density

hist(BankChurners$Customer_Age,freq = FALSE, col = "light blue", main="Umur Customer")
polygon(density(BankChurners$Customer_Age),border = "red") #Histogram&Density
```

```
hist(BankChurners$Customer_Age,freq = FALSE, col = "light blue", main="Umur Customer")
```

# *Box plot*

Fungsi ini digunakan untuk menapilkan *box plot* yang merupakan visualisasi dari penyebaran data (seperti kuartil, maksimum, minimum, dan pencilan). Penyebaran data yang ditampilkan dalam *box plot* ini adalah Customer_Age.

```
boxplot(BankChurners$Customer_Age, col= "light blue", main= "Umur Kustomer")
```

```
mean(BankChurners$Customer_Age)
```

```
heatmap(BankChurners$Customer_Age)
```
